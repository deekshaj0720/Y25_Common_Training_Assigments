## 3. Basic Image Functions

```python
# Write your code here
import cv2
import matplotlib.pyplot as plt 

img = cv2.imread('park.jpg')
#img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
grayscale = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

blur = cv2.GaussianBlur(grayscale,(7,7), cv2.BORDER_DEFAULT)

canny = cv2.Canny(blur,75,125)

plt.imshow(canny, cmap ='gray')
plt.axis("off")
plt.show()
```

## 4. Image Transformations
**Task:**
1. Create a function `translate(img, x, y)` to shift an image.
2. Create a function `rotate(img, angle, rotPoint=None)` to rotate an image.
3. Apply a flip transformation to `park.jpg`.

```python
# Write your code here
import cv2
import matplotlib.pyplot as plt 

img = cv2.imread('park.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

def translate(img,x,y):
    transMAT = np.float32([[1,0,x],[0,1,y]])
    dimensions = (img.shape[1],img.shape[0])
    return cv2.warpAffine(img,transMAT,dimensions)

def rotate(img, angle,rotPoint=None):
    (height,width) = img.shape[:2]

    if rotPoint is None:
        rotPoint =(width//2, height//2)

    rotMAT = cv2.getRotationMatrix2D(rotPoint,angle,1.0)
    dimensions = (width,height)

    return cv2.warpAffine(img,rotMAT,dimensions)

flip = cv2.flip(img, 0)

plt.imshow(flip)
plt.axis("off")
plt.show()
```

## 7. Smoothing and Blurring
**Task:**
1. Load `cats.jpg`.
2. Apply Averaging, Gaussian and Median Blurs

```python
# Write your code here
import cv2
import matplotlib.pyplot as plt 

img = cv2.imread('cat.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

#averaging
a_blur = cv2.blur(img, (7,7))

plt.imshow(a_blur)
plt.title('Averaging blur')
plt.axis("off")
plt.show()

#gaussian
g_blur = cv2.GaussianBlur(img, (7,7),0)

plt.imshow(g_blur)
plt.title('Gaussian blur')
plt.axis("off")
plt.show()

#median
m_blur = cv2.medianBlur(img, 7)

plt.imshow(m_blur)
plt.title('Median blur')
plt.axis("off")
plt.show()
```

## 8. Bitwise Operations
**Task:**
1. Create a $400 \times 400$ blank image.
2. Create a rectangle and a circle on separate blank images.
3. Perform Bitwise AND, OR, XOR, and NOT.

```python
# Write your code here
import cv2
import matplotlib.pyplot as plt 
import numpy as np 

blank = np.zeros((400,400,3),dtype ='uint8')
blank = cv2.cvtColor(blank, cv2.COLOR_BGR2RGB)

rectangle = cv2.rectangle(blank.copy(), (30,30),(370,370),(255,255,255),-1)
circle = cv2.circle(blank.copy(),(200,200),200,(255,255,255),-1)
plt.imshow(rectangle)
plt.axis("off")
plt.show()

plt.imshow(circle)
plt.axis("off")
plt.show()

#AND 
bitwise_and = cv2.bitwise_and(rectangle,circle)

plt.imshow(bitwise_and)
plt.title('AND')
plt.axis("off")
plt.show()

#OR
bitwise_or = cv2.bitwise_or(rectangle,circle)

plt.imshow(bitwise_or)
plt.title('OR')
plt.axis("off")
plt.show()

#XOR
bitwise_xor = cv2.bitwise_xor(rectangle,circle)

plt.imshow(bitwise_xor)
plt.title('XOR')
plt.axis("off")
plt.show()

#NOT
bitwise_not = cv2.bitwise_not(rectangle,circle)

plt.imshow(bitwise_not)
plt.title('NOT')
plt.axis("off")
plt.show()

```

## 9. Thresholding
**Task:**
1. Load `cats.jpg` and convert to grayscale.
2. Apply Simple Thresholding (Binary and Inverse).

```python
# Write your code here
import cv2
import matplotlib.pyplot as plt 

img = cv2.imread('cat.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#binary 
threshold, thresh = cv2.threshold(gray,100,255,cv2.THRESH_BINARY)
plt.imshow(thresh, cmap = 'gray')
plt.title('Binary threshold')
plt.axis("off")
plt.show()

#inverse
threshold, thresh_inv = cv2.threshold(gray,100,255,cv2.THRESH_BINARY_INV)
plt.imshow(thresh_inv, cmap = 'gray')
plt.title('Inverse threshold')
plt.axis("off")
plt.show()

```
